# 4. Outlier detection

```python
df.info()
```

```python
sns.boxplot(df['precipitation'])  #it has lots of outliers
```

```python
lower_limit=df['precipitation'].mean()-3*df['precipitation'].std()
upper_limit=df['precipitation'].mean()+3*df['precipitation'].std()
print(lower_limit,upper_limit)
```

```python
df.loc[(df['precipitation']>upper_limit)|(df['precipitation']<lower_limit)]
```

```python
new_df=df.loc[(df['precipitation']>upper_limit)|(df['precipitation']<lower_limit)]
print('length old data',len(df))
print('length new data',len(new_df))
```

```python
sns.boxplot(new_df['precipitation'])  #it has lots of outliers
```

# 6.Variable Transformation

```python
df['date']=df['date'].astype('category')
df['precipitation']=df['precipitation'].astype(int)
df['temp_max']=df['temp_max'].astype(int)
df['temp_min']=df['temp_min'].astype(int)
df['wind']=df['wind'].astype(int)
df['weather']=df['weather'].astype('category')
```
